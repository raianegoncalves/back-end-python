# BASE VENDAS

```python
# Carregar o arquivo CSV
vendas = pd.read_csv('vendas.csv')

# Converter a coluna 'date' para o formato de data
vendas['date'] = pd.to_datetime(vendas['date'])
```

1. Criar um gráfico de linhas das vendas ao longo do tempo usando Matplotlib

```python
# Criando o gráfico de linhas para a série temporal de vendas
plt.figure(figsize=(10, 6))
plt.plot(vendas['date'], vendas['sales'], color='blue', marker='o', linestyle='-', markersize=4)
plt.title('Vendas ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Vendas')
plt.grid(True)
plt.show()
```

2. Criar um gráfico de decomposição sazonal usando statsmodels

```python
# Realizando a decomposição sazonal da série temporal
decomposicao = seasonal_decompose(vendas['sales'], model='additive', period=12)

# Plotando a decomposição sazonal
decomposicao.plot()
plt.show()
```

3. Criar um gráfico interativo que permita a exploração dos dados de vendas ao longo do tempo usando Plotly

```python
# Criando o gráfico de linhas interativo para explorar as vendas ao longo do tempo
fig = px.line(vendas, x='date', y='sales', title='Vendas ao Longo do Tempo (Interativo)')
fig.update_xaxes(title_text='Data')
fig.update_yaxes(title_text='Vendas')
fig.show()
```
