Utilizando o arquivo: carros.csv

1- Faça o upload do CSV carros.csv
2- Print seu head com 10 linhas
3- Agora print as colunas e index do dataframe
4- Faça um sort decrescente no dataframe pelo estado com maior quantidade de carros
5- Descubra qual a proporção de carro para a população de cada estado e insira em uma nova coluna
6- Faça um dataframe com Estado e proporção de carro por habitante

```python
import pandas as pd
```

```python
df = pd.read_csv('carros.csv')
df.head(10) #Print seu head com 10 linhas
```

```python
df.columns #Agora print as colunas e index do dataframe
```

```python
df.index #Agora print as colunas e index do dataframe
```

```python
df.sort_values('Quant Carros', ascending = False) #Faça um sort decrescente no dataframe pelo estado com maior quantidade de carros
```

```python
#Descubra qual a proporção de carro para a população de cada estado e insira em uma nova coluna
proporcao_carro_estado = df['Quant Carros']/df['População do Estado']
proporcao_carro_estado
```

```python
df['Carros por habitante'] = proporcao_carro_estado
```

```python
df.head()
```

```python
#Faça um dataframe com Estado e proporção de carro por habitante
df_estado_hab = df[['Estado', 'Carros por habitante']]
```

AGREGALÇAO DE DADOS 

Utilizando o arquivo concurso.csv

7 - Quantas pessoas são do Estado do Piaui?
8 -  Qual a distribuição de escolaridade por região por região?
9 - Qual é a porcentagem de pessoas com deficiencia?

```python
df_concurso = pd.read_csv('dados_concurso.csv')
df_concurso.head(10) #Print seu head com 10 linhas
```

```python
#Quantas pessoas são do Estado do Piaui com count
df_concurso[df_concurso['Estado']== 'PI']['Número de Inscrição'].count()
```

```python
#Quantas pessoas são do Estado do Piaui com groupby
df_concurso[['Número de Inscrição', 'Estado']].groupby('Estado').count()
```

```python
#Qual a distribuição de escolaridade por Estado?
df_concurso[['Estado', 'Escolaridade', 'Número de Inscrição']].groupby(['Estado', 'Escolaridade']).count()
```

```python
#Qual é a porcentagem de pessoas com deficiencia?
df_concurso['Deficiência'].value_counts(normalize=True) *100 #mas como queremos porcentagem, então colocar normalize = true
```
