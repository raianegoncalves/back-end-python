4. SER ADVOGADO OU SER REPRESENTANTE DE VENDAS, QUAL FAZ DORMIR MENOS?

```python
#FILTRAR ESSAS PESSOAS
profissoes = df[df['Profissão'].isin(['Advogado(a)', 'Representante de Vendas'])]
profissoes.head(10)
```

```python
#CALCULAR MÉDIA DE SONO DE CADA PROFISSÃO
media_sono = profissoes.groupby('Profissão')['Duração do sono'].mean()
media_sono
```

```python
#VERIFICAR QUAL PROFISSÃO TEM A MENOR MÉDIA DE SONO
dorme_menos = media_sono.idxmin()
menor_media_sono = media_sono.min()
```

```python
print(f"A profissão que dorme menos, em média, é {dorme_menos} com uma média de {menor_media_sono:.2f} horas de sono.")
```

```python
#PLOTAR GRÁFICO
media_sono.plot(x= 'Duração do sono', y= 'Profissão', kind = 'bar')
plt.show()
```

8. QUEM TEM MAIOR PRESSÃO SANGUÍNEA MÉDIA, HOMENS OU MULHERES? (CONSIDERE A MÉDIA)

```python
#PRIMEIRO, VOU CRIAR UM SUBCONJUNTO SEGURO PARA FAZER UMA CÓPIA E NÃO DAR 'SettingWithCopyWarning'
subconjunto = df[['Pressão Sanguínea', 'Gênero']].copy()
```

```python
#SEPARAR A COLUNA PRESSÃO SANGUÍNEA EM DUAS
subconjunto[['Sistólica', 'Diastólica']] = subconjunto['Pressão Sanguínea'].str.split('/', expand=True)
subconjunto
```

```python
#CONVERTER O DTYPE DA COLUNA PRESSÃO SANGUÍNEA DE OBJECT PARA FLOAT
subconjunto['Sistólica'] = subconjunto['Sistólica'].astype(float)
subconjunto['Diastólica'] = subconjunto['Diastólica'].astype(float)
```

```python
#AGORA VOU CALCULAR A MÉDIA DA PRESSÃO SANGUÍNEA GERAL
subconjunto['Pressão Sanguínea Média'] = (subconjunto['Sistólica'] + subconjunto['Diastólica']) / 2
subconjunto
```

```python
#CALCULAR A MÉDIA POR GENÊRO
media_pressao = subconjunto.groupby('Gênero')['Pressão Sanguínea Média'].mean()
media_pressao
```

10. PESSOAS COM FREQUÊNCIA CARDÍACA > 70 DÃO + PASSOS QUE PESSOAS COM FREQUÊNCIA CARDÍACA =<70? (USE MÉDIA).

```python
#SEPARAR O GRUPO COM BASE EM FREQUÊNCIA
grupo_acima70 = df[df['Frequência cardíaca'] > 70] #grupo acima de 70
grupo_abaixo70 = df[df['Frequência cardíaca'] <= 70] #gtupo abaixo de 70
```

```python
#CALCULAR MÉDIA DE PASSOS
media_acima70 = grupo_acima70['Passos diários'].mean() #média acima de 70
media_abaixo70 = grupo_abaixo70['Passos diários'].mean() #média abaixo de 70
```

```python
#COMPARAÇÃO ENTRE AS MÉDIAS
if media_acima70 > media_abaixo70:
    print("Pessoas com frequência cardíaca > 70 dão mais passos.")
else:
    print("Pessoas com frequência cardíaca <= 70 dão mais passos ou têm médias iguais.")
```

```python
#RESULTADOS
print(f"Média de passos (FC > 70): {media_acima70}")
print(f"Média de passos (FC <= 70): {media_abaixo70}")
```
